```python
import tkinter
import tkintermapview
import random
import time

# create tkinter window
root_tk = tkinter.Tk()
root_tk.geometry(f"{800}x{600}")
root_tk.title("map_view_example.py")

# create map widget
map_widget = tkintermapview.TkinterMapView(root_tk, width=800, height=600, corner_radius=0)
map_widget.place(relx=0.5, rely=0.5, anchor=tkinter.CENTER)

# set current widget position and zoom
map_widget.set_position(39.9534791, 32.8731428)  # Paris, France
map_widget.set_zoom(5)

# set current widget position by address
map_widget.set_address("istanbul")

# print(dir(map_widget))
marker_ankara = map_widget.set_address("ankara", marker=True)
marker_ankara.set_text("ANKARA ðŸŽ‚")

marker_bursa = map_widget.set_address("bursa", marker=True)
marker_bursa.set_text("BURSA ðŸŽ‚")

marker_istanbul = map_widget.set_marker(41.0072359, 28.9752771, text="Ä°STANBUL")

# marker_berlin = map_widget.set_address("berlin", marker = True)

# set a path
# path_1 = map_widget.set_path([marker_berlin.position, marker_istanbul.position, marker_ankara.position, (38.7498443, 35.5008656) ])
# path_1.add_position(...)
# path_1.remove_position(...)
# path_1.delete()

color_list = ["red", "blue", "green", "pink", "light blue", "white", "black", "purple"]

def change_polygon_color():
    fill = random.choice(color_list)
    outline = random.choice(color_list)

    # function iÃ§erisinde yeni polygon Ã§iz ve boya
    polygon = map_widget.set_polygon([marker_istanbul.position,
                                    marker_ankara.position,
                                   marker_bursa.position],
                                   fill_color = fill,
                                   outline_color = outline,
                                   border_width = 10,
                                   command = polygon_click,
                                   name="poligon2")

def polygon_click(polygon):
    print(f"polygon clicked - text: {polygon.name}")
    
    for i in range(1):
        time.sleep(0.01)
        change_polygon_color()
    
    
polygon_1 = map_widget.set_polygon([marker_istanbul.position,
                                    marker_ankara.position,
                                   marker_bursa.position],
                                   fill_color = "yellow",
                                   outline_color = "green",
                                   border_width = 10,
                                   command = polygon_click,
                                   name="poligon1")

def add_marker_event(coords):
    print("Add marker:", coords)
    new_marker = map_widget.set_marker(coords[0], coords[1], text="new marker")
    

map_widget.add_right_click_menu_command(label="Add Marker",
                                        command=add_marker_event,
                                        pass_coords=True)


def left_click_event(coordinates_tuple):
    print("Left click event with coordinates:", coordinates_tuple)
    
map_widget.add_left_click_map_command(left_click_event)

polygon_list = []

# polygon yap fonksiyonu
def make_polygon(coords):
    print("ploygon listesi", polygon_list)
    new_marker = map_widget.set_marker(coords[0], coords[1], text= str(len(polygon_list)))
    polygon_list.append(coords)
    
    

map_widget.add_right_click_menu_command(label="make polygon",
                                        command = make_polygon,
                                        pass_coords = True)


def draw_polygon():
    print("draw a polygon")
    fill = random.choice(color_list)
    outline = random.choice(color_list)
    # function iÃ§erisinde yeni polygon Ã§iz ve boya
    polygon = map_widget.set_polygon(polygon_list,
                                   fill_color = fill,
                                   outline_color = outline,
                                   border_width = 10,
                                   command = polygon_click,
                                   name="poligon2")
    # write func call ediyoruz
    write()
    
map_widget.add_right_click_menu_command(label="draw polygon",
                                        command = draw_polygon,
                                        pass_coords = False)

# right click menÃ¼sÃ¼ ekleyin, eklediÄŸinizi menÃ¼ polygon_list clear() elemanlarÄ±nÄ± silecek

# dosyaya yazÄ± yazma
def write():
    try:
        dosya = open("coords.txt", "a") # a append demek yani eklemek demektir

        for i in polygon_list:
            dosya.write(str(i[0]) + "," + str(i[1]) + "\n") # burad tuple kullandÄ±k
            # str ye Ã§evirdik

        dosya.close()
    except:
        print("dosya yazma hatasÄ±")


tkinter.mainloop()
```
